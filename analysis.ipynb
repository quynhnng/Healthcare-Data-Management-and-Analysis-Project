```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#load data
df = pd.read_csv('processed_data.csv')

df.head(10)
```

```python
#plot the distribution of age groups
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='age')
plt.title('Distribution of Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

```python
#plot the distribution of readmissions
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='readmitted')
plt.title('Distribution of Readmissions')
plt.xlabel('Readmission')
plt.ylabel('Count')
plt.show()
```

```python
#plot the distribution of gender
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='gender')
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])

#plot the correlation matrix
plt.figure(figsize=(12, 10))
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
#relationship between age and readmission
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='age', hue='readmitted')
plt.title('Age Group vs. Readmission')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Readmitted')
plt.show()
```

```python
#distribution of primary diagnosis
plt.figure(figsize=(15, 8))
sns.countplot(data=df, y='diag_1', order=df['diag_1'].value_counts().index[:20])
plt.title('Top 20 Primary Diagnosis Codes')
plt.xlabel('Count')
plt.ylabel('Diagnosis Code')
plt.show()
```

```python
#distribution of time in hospital
plt.figure(figsize=(10, 6))
sns.histplot(df['time_in_hospital'], bins=20, kde=True)
plt.title('Distribution of Time in Hospital')
plt.xlabel('Time in Hospital (Days)')
plt.ylabel('Count')
plt.show()
```

```python
#Insulin Usage vs. Readmission
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='insulin', hue='readmitted')
plt.title('Insulin Usage vs. Readmission')
plt.xlabel('Insulin Usage')
plt.ylabel('Count')
plt.legend(title='Readmitted')
plt.show()
```
